```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

## Cleaning numerical data

```python
customer_df = pd.read_csv('files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
display(customer_df.head())
display(customer_df.shape)
display(customer_df.info())
```

```python
#convert the column names to lower case
customer_df.columns=customer_df.columns.str.lower()

#White spaces in column names should be replaced by _

customer_df.columns = customer_df.columns.str.replace(' ', '_') 
customer_df
```

```python
#Change effective to date column to datetime format.
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
customer_df
```

```python
#Create a NEW dataframe using only the numerical data called numerical_df.
numerical_df = customer_df.select_dtypes(include='number')
numerical_df
```

```python
numerical_df.nunique()
```

```python
#We want to split the numerical variables into descrete and continuous ones.
# When I look at the unique values I would suggest the limit to be 210 so there will be 3 continuous categories.

def label_numerical_columns(df):
    continuous_cols = []
    discrete_cols = []

    for col in df.columns:
        unique_values = df[col].nunique()
        if unique_values > 210:
            continuous_cols.append(col)
        else:
            discrete_cols.append(col)

    continuous_df = df[continuous_cols]
    discrete_df = df[discrete_cols]

    return continuous_df, discrete_df
```

```python
continuous_df, discrete_df = label_numerical_columns(numerical_df)
```

```python
continuous_df
```

```python
discrete_df
```

```python
# Because the total_claim_amount is our target variable and it is in the continuous dataframe, it makes sense to create the correlation matrix based on the orginal numerical_df.
# So we can check all numerical correlations.
correlations_matrix = numerical_df.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

```python
# We see that the discrete values despite monthly_premium_auto have a low correlation with total_claim_amount.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_discrete_var(df):
    for col in df.columns:
        sns.countplot(x=col, data=df)
        plt.title(f"Count Plot for {col}")
        plt.show()

def plot_continuous_var(df):
    for col in df.columns:
        sns.histplot(df[col], kde=True)
        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()
```

```python
plot_discrete_var(discrete_df)
```

```python
plot_continuous_var(continuous_df)
```

```python
# discrete variables: 
# monthly premium auto: there are way more people that pay a samll amount each month but the range is quite big.
# month since last claim: it can indicate the risk profile for a customer, you might want to check how to deal with the customers have a higher risk profile n detail.
# month since policy inception: Kind of evenly spread, this means we keep the customers we have for a while and we get new ones, too
# num of open complaints: Most of the customers do not have an open complaint.
# num of policies: Most poeple have 1 to 3 policies. Customers with more policies might be mostly corporate customers.
```

```python
# continous variables:
# CLV: We see quite big outliers here
# income: A lot of people have a really low income, but when we first looked at the data there were some incomes with 0, it could be that NaN Values once were filled with 0. 
# total claim amount: We see some outliers. But since this is our target I do not want to delete any outliers here, since the numbers still seem reasonable.
```

```python
# We want to look into the outliers more detailed

sns.boxplot(data=continuous_df)
```

```python
# I want to get rid of the outliers in customer lifetime value

iqr = np.percentile(continuous_df['customer_lifetime_value'],75) - np.percentile(continuous_df['customer_lifetime_value'],25)
upper_limit = np.percentile(continuous_df['customer_lifetime_value'],75) + 1.5*iqr
lower_limit = np.percentile(continuous_df['customer_lifetime_value'],25) - 1.5*iqr
continuous_df = continuous_df[(continuous_df['customer_lifetime_value']>lower_limit) & (continuous_df['customer_lifetime_value']<upper_limit)]
```

```python
sns.boxplot(data=continuous_df)
```

```python
# I want to remove the outliers from the main dataframe and the numeric one as well.

numerical_df = numerical_df[(numerical_df['customer_lifetime_value']>lower_limit) & (numerical_df['customer_lifetime_value']<upper_limit)]
customer_df = customer_df[(customer_df['customer_lifetime_value']>lower_limit) & (customer_df['customer_lifetime_value']<upper_limit)]
```

## Cleaning categorical data

```python
categorical_df = customer_df.select_dtypes(include = object)
#Let's drop the customer column. It has the type object but is more like an index column
categorical_df = categorical_df.drop(['customer'], axis=1)
categorical_df
```

```python
# Checking for NaN values

categorical_df.isna().sum()
```

```python
# Luckily we do not have NaN values. 
```

```python
# Checking for unqiue values

for col in categorical_df.columns:
    unique_values = categorical_df[col].unique()
    print(f"Column: {col}")
    print(f"Unique values: {unique_values}")
```

```python
# It would be nice to directly see the value count, too.

for col in categorical_df.columns:
    unique_values = categorical_df[col].unique()
    print(f"Column: {col}")
    print(f"Unique values:")
    for value in unique_values:
        count = categorical_df[col].value_counts()[value]
        print(f" - {value}:{count}")
    print()
```

```python
# On the first sight that looks all quite reasonable. policy_type and policy seem to be quite similiar so that would be something whereI would have a closer look on.
```

```python
#Check dtypes. Do they all make sense as categorical data?
categorical_df.dtypes
```

```python
#Yes, it does make sense to keep them as categorical data.
```

```python
#Check for alpha + numeric data. Decide how to clean it and do it now.
#There are two columns policy and renew_offer_type. With policy we have to check the meaning of it first, so for now I would leave it like that
#renew_offer_type we can make the categories easier if we leave out the word "Offer". Than we kind of already have an enocded category.

categorical_df['renew_offer_type'] = categorical_df['renew_offer_type'].str.replace("Offer","")
categorical_df['renew_offer_type'].unique()
```

```python
#Would you choose to do anything else to clean or wrangle the categorical data? Comment your decisions and do it now.
#As we did before in a Lab we can generate a category Luxury for vehicle_class.
#Actually vehicle_class and vehicle_size are quite redundant. The task did not really say anything about dropping columns but I think we can keep that in mind.

categorical_df['vehicle_class'] = np.where(categorical_df['vehicle_class'].isin(['Sports Car','Luxury SUV','Luxury Car']) , 'Luxury', categorical_df['vehicle_class'])
categorical_df['vehicle_class'].value_counts()
```

```python
#I also decide to merge medical leave and disabled to health_related_absence. If you get the status on medical leave it seems to be something permanent that keeps you from working and therefor it seems kind of the same as having a disability.
#Retired is also a small category but since that has to do something with age I gues for an insurance company this might have a high correlation so i do not want to group it, too.

categorical_df['employmentstatus'] = np.where(categorical_df['employmentstatus'].isin(['Medical Leave','Disabled']) , 'health related absence', categorical_df['employmentstatus'])
categorical_df['employmentstatus'].value_counts()
```

```python
# In the education column I could merge Master and Doctor. I would argue that it kind of stands for the same intelectual background and I would expect a high salary for both categories.

categorical_df['education'] = np.where(categorical_df['education'].isin(['Master','Doctor']) , 'Master or Doctor', categorical_df['education'])
categorical_df['education'].value_counts()
```

```python
#Compare policy_type and policy. What information is contained in these columns. Can you identify what is important?
#I can not really see the benefit of the column policy. It adds more subcategories but does not add value trough this. That is why I would jsut drop this column.

categorical_df = categorical_df.drop(['policy'], axis=1)
categorical_df
```

```python
categorical_df.to_csv('categorical.csv', index=False)
```
